# Tennis Ace

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and investigate the data here:

stats = pd.read_csv('tennis_stats.csv')

stats.head()
```

```python
stats.describe(include='all')
```

```python
plt.scatter(stats[['BreakPointsFaced']], stats[['Wins']], alpha=0.4)
# Create x-axis label and y-axis label
plt.xlabel("Break Poinst Faced")
plt.ylabel("Wins")
 
# Create a title
plt.title("Break Poinst Faced VS Winings")
plt.show()
```

```python
plt.scatter(stats[['BreakPointsOpportunities']], stats[['Winnings']], alpha=0.4)
# Create x-axis label and y-axis label
plt.xlabel("Break Points Opportunities")
plt.ylabel("Winnings")
 
# Create a title
plt.title("Break Points Opportunities VS Winings")
plt.show()
```

#### Linear Regression

```python
x = stats[['BreakPointsOpportunities']]
y = stats[['Winnings']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=6)

lr = LinearRegression()

lr.fit(x_train, y_train)

wins_predicted = lr.predict(x_test)

print("Train score: " + str(lr.score(x_train, y_train)))

print("Test score: " + str(lr.score(x_test, y_test)))

# Create a scatter plot
plt.scatter(y_test, wins_predicted, alpha=0.4)
 
# Create x-axis label and y-axis label
plt.xlabel("Actual Winnings")
plt.ylabel("Predicted Winnings")
 
# Create a title
plt.title("Actual Winnings vs Predicted Winnings")
 
# Show the plot
plt.show()
```

#### Multiple Linear Regression

```python
x = stats[['BreakPointsOpportunities', 'Aces']]
y = stats[['Winnings']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=6)

mlr = LinearRegression()

mlr.fit(x_train, y_train)

outcome_predicted = mlr.predict(x_test)

print("Train score: " + str(mlr.score(x_train, y_train)))

print("Test score: " + str(mlr.score(x_test, y_test)))

# Create a scatter plot
plt.scatter(y_test, outcome_predicted, alpha=0.4)
 
# Create x-axis label and y-axis label
plt.xlabel("Actual Winnings")
plt.ylabel("Predicted Winnings")
 
# Create a title
plt.title("Actual Winnings vs Predicted Winnings")
 
# Show the plot
plt.show()
```
